```python
import gym
import numpy as np
import matplotlib.pyplot as plt

from data.sequence import SequenceDataset
from data.d4rl import get_dataset
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
for e_idx, env in enumerate(["hopper", "walker2d", "halfcheetah"]):
    for q_idx, quality in enumerate(["medium-expert", "medium", "medium-replay"]):
        data = get_dataset(
            gym.make(f"{env}-{quality}-v2"),
            max_traj_length=1000,
            include_next_obs=False,
            termination_penalty=-100.0,
        )
        sequence_dataset = SequenceDataset(
            data,
            horizon=20,
            max_traj_length=1000,
            include_returns=True,
            include_cost_returns=False,
            normalizer="LimitsNormalizer",
            use_padding=True,
        )

        returns = []
        for idx in range(len(sequence_dataset)):
            returns.append(sequence_dataset[idx]["returns"].squeeze())
        returns = np.array(returns)

        axes[e_idx][q_idx].hist(returns, bins=100)
        axes[e_idx][q_idx].set_title(f"{env}-{quality}")

plt.tight_layout()
fig.savefig("save_figs/d4rl_h20.png")
```
